```python
import polars as pl
import pandas as pd # It is like polars but better
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_polars = pl.DataFrame({
    'Category': ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C'],
    'Values': [1.2, 2.3, 2.1, 3.5, 3.8, 3.1, 4.5, 4.2, 4.8]
})

data = data_polars.to_pandas()

data
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='Category', y='Values', data=data)

plt.show()
```
